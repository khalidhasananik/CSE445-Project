```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'dataset/dataset.csv'
data = pd.read_csv(file_path)

# handle missing values with mode for whole dataset

data.ffill(axis=0, inplace=True)


# Define the months
months = ['January', 'February', 'March', 'April', 'May', 'June',
          'July', 'August', 'September', 'October', 'November', 'December']

# Calculate the total dengue cases across all districts and years for each month
monthly_trend = data[months].sum()

# Plot the overall monthly trend
plt.figure(figsize=(10, 6))
plt.plot(months, monthly_trend, marker='o', linestyle='-', color='lightblue', label='Monthly Dengue Cases')

# Add titles and labels
plt.title('Overall Monthly Trend of Dengue Cases', fontsize=16)
plt.xlabel('Months', fontsize=14)
plt.ylabel('Total Dengue Cases', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.legend(fontsize=12)
plt.tight_layout()

# Save the plot
plt.savefig('overall_monthly_trend.png')

# Show the plot
plt.show()
```
